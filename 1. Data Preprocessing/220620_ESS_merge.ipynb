```python
import os
import pandas as pd
import datetime as dt
import numpy as np
import time
from datetime import tzinfo, timedelta, datetime
```

```python
## 폴더 내 엑셀 파일 list화

file_format = ".csv"   ##파일형식
file_path = '/Users/jihyeon/Downloads/0603/PCS3_TREND'  ## 파일 경로
file_list = [f"{file_path}/{file}" for file in os.listdir(file_path) if file_format in file]
file_list
```

```python
merge_df = pd.DataFrame()

for file_name in file_list:
    ## 차례대로 다읽어와
    file_df = pd.read_csv(file_name, encoding='euc-kr')
    
    ## 컬럼명 변경
    file_df = file_df.rename(columns={'Unnamed: 0':'DateTime'})
    
    ## DateTime 괄호 없애고 datetime 타입으로 변경
    file_df['DateTime'] = file_df['DateTime'].str.replace("[","")
    file_df['DateTime'] = file_df['DateTime'].str.replace("]","")
    file_df['DateTime'] = pd.to_datetime(file_df['DateTime'], format="%Y/%m/%d %H:%M:%S", errors='coerce')
    
    ## 합쳐 날짜순으로 정렬해
    merge_df = pd.concat([merge_df, file_df])
    merge_df = merge_df.sort_values(by='DateTime', axis=0)
```

```python
## 합친 파일 csv파일로 저장 (index 없어)

merge_df.to_csv("/Users/jihyeon/Downloads/0603/PCS2_TREND/PCS2_3.csv", index = False, encoding='euc-kr')
```
